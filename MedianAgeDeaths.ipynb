```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

data = pd.read_csv('medianage.csv', index_col = "NEIGHBORHOOD")
data = data.filter(regex = "TOTAL MD AGE AT DEATH")
data = data.dropna()
data = data[data["TOTAL MD AGE AT DEATH"] > 0]
data = data.drop_duplicates()
data = data.nlargest(10, 'TOTAL MD AGE AT DEATH')
data = data.sort_values('TOTAL MD AGE AT DEATH')

plt.style.use('fivethirtyeight')
data.plot(kind='barh', legend=False)
plt.title("Highest Median Age at Death in Pittsburgh Neighborhoods")
plt.xlabel("Median Age at Death")
plt.ylabel("Neighborhood")
```

My dataset describes the median age deaths in Pittsburgh neighborhoods. The sub-metric used is "health/healthiest neighborhoods," because generally higher median age deaths correlate with better health and living conditions of an area. Median is used rather than mean because age deaths may include skewed data values or outliers that may inaccurately represent what an average value may look like. In the data visaulization, I narrowed it down to the top 10 highest median age deaths in Pittsburgh neighborhoods. From the data, it is evident that there is not much difference among the top ten neighborhoods. They all are above 80 in their median age at death and are only separated by small margins. Therefore, if I were only to consider my metric, these neighborhoods would have a similar high quality of life. However for the purpose of choosing the "best" neighborhood, Squirrel Hill South would be the best neighborhood as it has the highest median age at death.
