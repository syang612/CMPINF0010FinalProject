```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt


trashData = pd.read_csv("smarttrash.csv")

smartTrash = trashData['neighborhood'].value_counts().head(20)
print(smartTrash)

plt.figure(figsize=(15, 8))
smartTrash.plot(kind='bar', color='green', edgecolor='black')
plt.title('Number of Smart Trash Containers in Each Neighborhood', fontsize=16)
plt.xlabel('Neighborhood', fontsize=12)
plt.ylabel('Number of Smart Trash Containers', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.tight_layout()

plt.show()
```

This graph shows the number of Smart Trash Containers per neighborhood in Pittsburgh. The sub-metric of this data set is "cleaniliness". I arranged it so it would show the top 20 neighborhoods with the highest amount of smart trash containers. The "best" neighborhood according to my metric would be Shadyside because of its 76 smart trash containers being the highest . My dataset is important because it demonstrates the cleanliness levels of each neighborhood in Pittsburgh, which majorly contributes to overrall health and quality of life of each neighborhood to determine which is the best.
