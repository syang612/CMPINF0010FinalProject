# Number of Crosswalks in Each Neighborhood

Through this project, I look at the amount of crosswalks within each county to determine which neighborhood is the safest for pedestrians. An abundance of crosswalks enhances safety by encouraging pedestrians to cross at designated areas, reducing the risk of jaywalking and potential accidents. They increase driver awareness by signaling pedestrian presence and help manage traffic flow by reducing conflict points between vehicles and pedestrians. Crosswalks also improve accessibility for vulnerable groups, calm traffic with accompanying measures like signals or speed bumps, and promote walkability, reducing congestion and collisions. To be effective, they must be well-placed, maintained, and supported by additional safety features such as lighting and pedestrian signals.

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt


crosswalk = pd.read_csv("crosswalks.csv")

numberOfCrosswalks = crosswalk['neighborhood'].value_counts()

plt.figure(figsize=(20, 6))
numberOfCrosswalks.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Number of Crosswalks in Each Neighborhood', fontsize=16)
plt.xlabel('Neighborhood', fontsize=12)
plt.ylabel('Number of Crosswalks', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.tight_layout()

plt.show()

```

This dataset highlights the number of crosswalks in each neighborhood. Based on the graph, it can be concluded that the Central Business District is the best neighborhood for pedestrian safety, as it has 400 crosswalks. Following closely behind are South Side Flats and Squirrel Hill South, each with approximately 250 crosswalks. The abundance of crosswalks is important in a neighborhood because it helps drivers anticipate pedestrians, encouraging them to yield and allowing for safer crossings with a reduced risk of accidents.
