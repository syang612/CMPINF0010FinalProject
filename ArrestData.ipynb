```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
arrestData = pd.read_csv("arrest.csv")
arrestData.head()
arrestascending = arrestData['INCIDENTNEIGHBORHOOD'].value_counts(ascending=True).to_frame()
arrestascending.reset_index(inplace=True) 
arrestascending.columns = ['Neighborhood', 'Number of Arrests'] 
arrestascending.head(10)

plt.figure(figsize=(12, 6))
plt.bar(arrestascending['Neighborhood'].head(10),
        arrestascending['Number of Arrests'].head(10),
        color="blue", edgecolor='black')
plt.title("Arrest Data")
plt.xlabel("Neighborhood", fontsize=14)
plt.ylabel("Number of Arrests", fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```
