The abundance of crosswalks is important in a neighborhood because it helps drivers anticipate pedestrians, encouraging them to yield and allowing for safer crossings with a reduced risk of accidents.

# Median Age at Death in Pittsburgh Neighborhoods

My dataset describes the median age deaths in Pittsburgh neighborhoods. The sub-metric used is "healthiest neighborhoods," because generally higher median age deaths correlate with healthier lifestyles and better living conditions of an area. Median is used rather than mean because this dataset may include skewed data values or outliers that may inaccurately represent what an average value may look like, and using median means being less affected by these outliers. 

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

data = pd.read_csv('medianage.csv', index_col = "NEIGHBORHOOD")
data = data.filter(regex = "TOTAL MD AGE AT DEATH")
data = data.dropna()
data = data[data["TOTAL MD AGE AT DEATH"] > 0]
data = data.drop_duplicates()
data = data.nlargest(15, 'TOTAL MD AGE AT DEATH')
data = data.sort_values('TOTAL MD AGE AT DEATH')
print(data)

plt.style.use('fivethirtyeight')
data.plot(kind='barh', legend=False, color='orange', edgecolor='black')
plt.title("Highest Median Age at Death in Pittsburgh Neighborhoods")
plt.xlabel("Median Age at Death")
plt.ylabel("Neighborhood")
```

In the data visaulization, I narrowed it down to the top 15 highest median age deaths in Pittsburgh neighborhoods. In my code, I filtered through the data through methods such as ".drop_duplicates()" and ".dropna()" to "clean" the data and make sure that the data was an accurate representation of the neighborhoods with the highest median age at death. Finally, I used matplotlib to graph in the data in a bar chart. From the data, it is evident that there is not much difference among the top ten neighborhoods. They all are on or above 80 in their median age at death and are only separated by tenths of a margin. Therefore, if I were only to consider my metric, these neighborhoods would have a similar high quality of life. However for the purpose of choosing the "best" neighborhood, Squirrel Hill South would be the best neighborhood as it has the highest median age at death.

# Combined Data

```python
import pandas as pd
import matplotlib.pyplot as plt

crosswalk = pd.read_csv("crosswalks.csv")
crosswalk_counts = crosswalk['neighborhood'].value_counts()

trashData = pd.read_csv("smarttrash.csv")
smart_trash_counts = trashData['neighborhood'].value_counts()

median_age_data = pd.read_csv('medianage.csv')
median_age = median_age_data['NEIGHBORHOOD'].value_counts()


combined_data = pd.DataFrame({
    "Crosswalks": crosswalk_counts,
    "Smart Trash": smart_trash_counts,
    "Median Age at Death": median_age
})

combined_data['Total'] = combined_data.sum(axis=1)
top_neighborhoods = combined_data.sort_values(by="Total", ascending=False).head(20)

top_neighborhoods.drop(columns='Total').plot(
    kind="barh",
    stacked=True,
    color=["skyblue", "green", "orange"],
    edgecolor="black",
    figsize=(12, 8)
)
plt.title("Top Pittsburgh Neighborhoods: Crosswalks, Smart Trash, and Median Age at Death", fontsize=16)
plt.xlabel("Total Count")
plt.ylabel("Neighborhood")
plt.legend(title="Type", fontsize=12)
plt.tight_layout()

plt.show()
```

For a complete data visualization, we decided upon a stacked bar graph. Since all of our datasets resulted in the "highest" amount being the "best" in each of our respective sub-metrics, we concluded that the neighborhood with the highest count including all our metrics would be the "best" neighborhood. The code above demonstrates the creation of a dataframe with columns containing the data values of each of our three datasets. We combined the data values into a "Total" column, and displayed the top 20 highest data values in 'Total' in a plotted stacked bar graph. As the data visualization displays, we conclude that the best neighborhood in Pittsburgh is the **Central Business District or Downtown**. This conclusion is backed by our reasoning that since it is the highest combined total of our data values collected from each of our sub-metrics, then it is the has the highest quality of life.
