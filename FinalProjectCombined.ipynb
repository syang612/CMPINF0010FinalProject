**Introduction**

Our project focuses on identifying the best neighborhood to live in by analyzing three key factors: the average median age of death, the number of crosswalks, and the number of smart trash container. Our overall goal is to determine the best neighborhood through the metric of "highest quality of life" by evaluating through the sub-metrics of healthiness, cleanliness, and safety of each community.

The median age of death dataset examines the average age at which individuals in each neighborhood pass away. A higher median age of death is an indicator of better overall health and living conditions within a community. The crosswalks dataset analyzes the number of crosswalks in each neighborhood, which reflects the communityâ€™s commitment to pedestrian and driver safety. A higher number of crosswalks promotes safer streets by encouraging drivers to yield to pedestrians, thereby reducing the risk of accidents. Lastly, the smart trashcan dataset highlights the number of smart trashcans in each neighborhood, reinforcing the idea that neighborhoods with an abundance of smart trashcans tend to be cleaner and more desirable places to live.

To ensure a focused analysis, we narrowed our study to neighborhoods within the city of Pittsburgh. This allowed us to examine each neighborhood in greater detail. However, some of our datasets contained conflicting information, which posed challenges when combining and synthesizing the data.


```python
import pandas as pd
import matplotlib.pyplot as plt

crosswalk = pd.read_csv("crosswalks.csv")
crosswalk_counts = crosswalk['neighborhood'].value_counts()

trashData = pd.read_csv("smarttrash.csv")
smart_trash_counts = trashData['neighborhood'].value_counts()

median_age_data = pd.read_csv('medianage.csv')
median_age = median_age_data['NEIGHBORHOOD'].value_counts()


combined_data = pd.DataFrame({
    "Crosswalks": crosswalk_counts,
    "Smart Trash": smart_trash_counts,
    "Median Age at Death": median_age
})

combined_data['Total'] = combined_data.sum(axis=1)
top_neighborhoods = combined_data.sort_values(by="Total", ascending=False).head(20)

top_neighborhoods.drop(columns='Total').plot(
    kind="barh",
    stacked=True,
    color=["skyblue", "green", "orange"],
    edgecolor="black",
    figsize=(12, 8)
)
plt.title("Top Pittsburgh Neighborhoods: Crosswalks, Smart Trash, and Median Age at Death", fontsize=16)
plt.xlabel("Total Count")
plt.ylabel("Neighborhood")
plt.legend(title="Type", fontsize=12)
plt.tight_layout()

plt.show()
```

For a complete data visualization, we decided upon a stacked bar graph. Since all of our datasets resulted in the "highest" amount being the "best" in each of our respective sub-metrics, we concluded that the neighborhood with the highest count including all our metrics would be the "best" neighborhood. The code above demonstrates the creation of a dataframe with columns containing the data values of each of our three datasets. We combined the data values into a "Total" column, and displayed the top 20 highest data values in 'Total' in a plotted stacked bar graph.

As the data visualization displays, we conclude that the best neighborhood in Pittsburgh is the **Central Business District or Downtown**. This conclusion is backed by our reasoning that since it is the highest combined total of our data values collected from each of our sub-metrics, then it is the has the highest quality of life.
